````python
# Answer Code
def play(sy, sx, N):
	y, x = sy, sx
	visited = [[0] * N for _ in range(N)]
	visited[y][x] = 1

	notEnd = True

	while notEnd:
		cnt = count[y][x]
		dy, dx = command[y][x]

		for _ in range(cnt):
			y = (y + dy) % N
			x = (x + dx) % N

			if visited[y][x]:
				notEnd = False
				break
			visited[y][x] = 1

	return sum([sum(i) for i in visited])

N = int(input())
Rg, Cg = map(int, input().split())
Rp, Cp = map(int, input().split())

Rg -= 1
Cg -= 1
Rp -= 1
Cp -= 1

arr = [list(input().split()) for _ in range(N)]

count = [[0] * N for _ in range(N)]
command = [[None] * N for _ in range(N)]
direction = {"L" : [0, -1], "R" : [0, 1], "U" : [-1, 0], "D" : [1, 0]}

for i in range(N):
	for j in range(N):
		temp = arr[i][j]
		count[i][j] = int(temp[:-1])
		key = temp[-1]
		command[i][j] = direction[key]

scoreG = play(Rg, Cg, N)
scoreP = play(Rp, Cp, N)

if scoreG > scoreP:
	print("goorm", scoreG)
else:
	print("player", scoreP)
````

- 예를 들어, 보드의 크기가 5라고 가정하고, 현재 말의 위치가 (4, 4)이고 이동 방향이 "R"이라면, 말은 현재 위치에서 오른쪽으로 이동하려고 합니다. 하지만 보드의 오른쪽 경계를 넘어가면, 남은 이동 횟수만큼을 반대편에서 이동하게 되어 (0, 4)로 이동합니다. 이것이 보드를 벗어난 경우의 동작입니다.

- 문제에서는 보드판의 모서리에 부딪히면 x,y 좌표가 반대로만 간다고 되어있어서, 그렇게 이해하고 풀었는데, 문제에 오류가 너무 많았다.

````python
# 게임을 시뮬레이션할 함수를 선언합니다. 시작하는 좌표인 sy, sx와 보드의 크기 N을 전달해줍니다.
def play(sy, sx, N):
	# 현재 위치를 담을 y, x를 선언합니다.
	y, x = sy, sx
	# 방문 배열을 선언하고, 시작 좌표에 대해 방문처리를 할게요.
	visited = [[0] * N for _ in range(N)]
	visited[y][x] = 1

	# 게임의 종료 조건을 충족할 때까지 시뮬레이션하기 위해 while문을 사용할거에요. 변수를 하나 선언할게요.
	notEnd = True

	# 게임이 종료되지 않은 동안 while문을 돌립니다.
	while notEnd:
		# 현재 위치에서의 count값과 방향을 불러옵니다.
		cnt = count[y][x]
		dy, dx = command[y][x]

		# cnt 만큼 dy, dx를 이용해 한 방향으로 이동합니다.
		for _ in range(cnt):
			y += dy

			# 범위를 벗어나는 경우에 대해 처리해줍니다. 오른쪽 끝으로 벗어나면 N, 왼쪽 끝으로 벗어나면 -1이겠죠.
			if y == N:
				y = 0
			elif y == -1:
				y = N - 1

			x += dx

			if x == N:
				x = 0
			elif x == -1:
				x = N - 1

			# 범위 처리를 완료했습니다. 만약 (y, x) 위치를 방문했다면 게임을 종료합니다.

			if visited[y][x]: # 정수값의 경우, 0 이외에는 True이므로 이렇게 써도 됩니다.
				notEnd = False
				break

			# 방문하지 않았었다면 방문처리를 할게요.
			visited[y][x] = 1

		# while문이 종료되면 여기로 올 것입니다. 게임에서 말하는 점수는 방문한 서로 다른 칸의 개수입니다.
		# 저희는 visited 배열에서 방문한 칸을 1로 지정했습니다. 즉, visited의 총합이 구하는 값입니다.

		return sum([sum(i) for i in visited])
````

### pass VS continue
```python
for i in somelist:
  if not element:
    pass
    print('a')

for k in somelist:
  if not element:
    continue
```
1. pass는 단순히 실행할 코드가 없다는 것을 의미 -> 뒤에 코드가 남아있다면 해당 코드를 실행시킴
2. continue는 뒤에 코드가 있든 말든 다음 순번의 loop를 돌도록 강제하는 것을 의미
3. pass와는 다르게 continue에서는 바로 다음 순번의 loop를 돈다.
