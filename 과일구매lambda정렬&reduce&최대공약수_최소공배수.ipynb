````python
# My Code
N, K = map(int,input().split())
poman = [0.1 for _ in range(N)]
price = [0.1 for _ in range(N)]

for i in range(N):
	pri,pom = map(int, input().split())
	poman[i] = pom
	price[i] = pri

natgae = [int(x / y) for x, y in zip(poman, price)]
bg = list(zip(natgae,price))
m = sorted(bg, reverse = True)
totalpoman = 0

for i in m:
	nat = i[0]
	pr = i[1]
	if K // pr >= 1:
		K = K - pr
		totalpoman += pr * nat
	elif K // pr < 1:
		totalpoman += K * nat
		break

print(totalpoman)
````

#### **리스트에 정렬 key 사용(sort, sorted)**
```python
a = [(1, 2), (5, 1), (0, 1), (5, 2), (3, 0)]
```
- key 인자를 정하지 않은 기본적인 sort에선, 튜플 순서대로 우선순위 기본 할당
```python
# 앞의 인자로 정렬됨
b = sorted(a)
b = [(0, 1), (1, 2), (3, 0), (5, 1), (5, 2)]
```
- key 인자에 함수를 넘겨주면 우선순위가 정해짐.

```python
c = sorted(a, key = lambda x : x[0])

c = [(0, 1), (1, 2), (3, 0), (5, 1), (5, 2)]

d = sorted(a, key = lambda x : x[1])

d = [(3, 0), (5, 1), (0, 1), (1, 2), (5, 2)]
```
- 비교할 아이템이 요소가 복수 개일 경우, 튜플로 우선순위를 정해줄 수 있다.

- -를 붙이면, 현재와 반대차순으로 정렬된다.

```python
e = sorted(a, key = lambda x : (x[0], -x[1]))
=> [(0, 1), (1, 2), (3, 0), (5, 2), (5, 1)]

f = sorted(a, key = lambda x : -x[0])
=> [(5, 1), (5, 2), (3, 0), (1, 2), (0, 1)])
```

- 뒤에 있는 문자를 기준으로 정렬하는 것도 가능
```python
s = ['2 A', '1 B', '4 C', '1 A']
s.sorted(s, key=lambda x: (x.split()[1], x.split()[0]))
=> ['1 A', '2 A', '1 B', '4 C']
```

````python
# Answer Code
N, K = map(int, input().split())
arr = [list(map(int, input().split())) for _ in range(N)]

arr.sort(key=lambda x : x[1] // x[0])

result = 0

while K and arr:
	P, C = arr.pop()

	if K >= P:
		result += C
		K -= P
	else:
		result += C // P * K
		K = 0
# arr 뿐만 아니라 K(가지고 있는 돈)도 0이 될때를 고려한다.
print(result)
````
